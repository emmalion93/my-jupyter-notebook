```python
# Single Layer Calculator

# Function retrieves dE/dx data from json dedx.json for associated target and ion
# Requires ion and target retrieved from other functions
# Returns list of dE/dx data for the given ion/target pair
def getDeDx(ion, target):
    # Current implementation: Grabbing json files from current user's directory by stepping down 2 levels (accounting for dir 'RUN' once called by Notebook).
    # Future implementation: Grab information directly from SIT
    f = open(os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir), "dedx.json"))
    dedxdata = json.load(f)
    dedxkey = ion.upper() + " " + target + " dE/dx"
    return dedxdata[dedxkey]


# Function retrieves XY Straggling (Lateral) from latstrag.json for associated target and ion
# Requires ion and target retrieved from other functions
# Returns list of lateral straggling data for the given ion/target pair
def getLatStrag(ion, target):
    f = open(os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir), "latstrag.json"))
    latstragdata = json.load(f)
    latstragkey = ion.upper() + " " + target + " LatStrag"
    return latstragdata[latstragkey]


# Function retrieves Z Straggling (Longitudinal) from longstrag.json for associated target and ion
# Requires ion and target retrieved from other functions
# Returns list of lateral straggling data for the given ion/target pair
def getLongStrag(ion, target):
    f = open(os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir), "longstrag.json"))
    longstragdata = json.load(f)
    longstragkey = ion.upper() + " " + target + " LongStrag"
    return longstragdata[longstragkey]


# Function retrieves Range data from range.json for associated target and ion
# Requires ion and target retrieved from other functions
# Returns list of range data for the given ion/target pair
def getRange(ion, target):
    f = open(os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir), "range.json"))
    rangedata = json.load(f)
    rangekey = ion.upper() + " " + target + " Range"
    return rangedata[rangekey]


# Function retrieves Energy data from energy.json for associated target and ion
# Requires ion and target retrieved from other functions
# Returns list of energy data for the given ion/target pair
def getEnergy(ion):
    f = open(os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir), "energy.json"))
    energydata = json.load(f)
    return energydata[ion.upper()]


# Function takes user input for a given energy, validates the input and finds the closest smallest match in the
# energy data Takes in energy list retrieved from other function Returns a list in the format of [index of item,
# closest match, input given]
def findInEnergy(energy):
    length = len(energy)
    i = 0
    while i < length:
        if energy[i] <= input_energy < energy[i + 1]:
            closestEnergy = energy[i]
            indexof = i
            energywindex = [indexof, closestEnergy, input_energy]
            i = length
        i += 1
    return energywindex


# Function takes in range data gathered from other function, takes user input, validates, and finds the closest
# smallest match to input Returns a list in the format of [index of item, closest match, input given]
def findInRange(range_mm):
    length = len(range_mm)
    i = 0
    while i < length:
        if range_mm[i] <= input_range < range_mm[i + 1]:
            closestRange = range_mm[i]
            indexof = i
            rangewindex = [indexof, closestRange, input_range]
            i = length
        i += 1
    return rangewindex


# Function takes in dE/dx data as list gathered from other function, takes user input, validates and finds closest
# smallest match Returns a list in the format of [index of item, closest match, given input]
def findInDedx(dedx):
    dedxwindex = []
    length = len(dedx)
    i = 0
    while i < length:
        if dedx[i] > input_dedx >= dedx[i + 1]:
            closestDedx = dedx[i + 1]
            indexof = i + 1
            dedxwindex = [indexof, closestDedx, input_dedx]
            i = length
        i += 1
    return dedxwindex


# Function takes gathered data for dE/dx input and full dE/dx data list
# Max value of dE/dx found, used to find difference for calculations
# Returns a list in the format of [matching difference, index of difference for other lists]
def dedxdiff(dedxwindex, dedx):
    difference = 0
    maximum = max(dedx)
    maxdiff = maximum - dedxwindex[2]
    maxindex = dedx.index(maximum)
    diffdict = {}
    diffvalues = []
    finaldiff = []
    i = maxindex
    while maxindex <= i < len(dedx):
        diffdict[maximum - dedx[i]] = i
        i += 1
    for keys in diffdict:
        diffvalues.append(keys)
    i = 0
    length = len(diffvalues)
    while i < length - 1:
        if diffvalues[i] <= maxdiff < diffvalues[i + 1]:
            difference = diffvalues[i]
            i = length - 1
        else:
            i += 1
    finaldiff.append(difference)
    finaldiff.append(diffdict[difference])
    return finaldiff

# Multi-Layer functions

# Prompts user to select target layers and thickness
# Returns a dictionary
def setLayers(length):
    layerdict = {}

    layerdict["Layer 1"] = [layer1_material, layer1_thickness]
    layerdict["Layer 2"] = [layer2_material, layer2_thickness]
    layerdict["Layer 3"] = [layer3_material, layer3_thickness]
    layerdict["Layer 4"] = [layer4_material, layer4_thickness]
    layerdict["Layer 5"] = [layer5_material, layer5_thickness]
    layerdict["Layer 6"] = [layer6_material, layer6_thickness]
    layerdict["Layer 7"] = [layer7_material, layer7_thickness]
    layerdict["Layer 8"] = [layer8_material, layer8_thickness]
    layerdict["Layer 9"] = [layer9_material, layer9_thickness]
    layerdict["Layer 10"] = [layer10_material, layer10_thickness]

    return layerdict

# Finds closest match to energy and returns in list [index, match, input]
def getEnergyIndex(energylist, energyin):
    energymatch = []
    for i in range(0, len(energylist) - 1):
        if energylist[i] <= energyin < energylist[i + 1]:
            energymatch.append(i)
            energymatch.append(energylist[i])
            energymatch.append(energyin)
    return energymatch


# Get the matching index for the closest match to the passed in range value
# Returns a list in the format [index, match, index up]
def getRangeIndex(rangelist, resrange):
    resrangematch = []
    for i in range(0, len(rangelist) - 1):
        if rangelist[i] <= resrange < rangelist[i + 1]:
            resrangematch.append(i)
            resrangematch.append(rangelist[i])
            resrangematch.append(rangelist[i + 1])
    if len(resrangematch) == 0:
        resrangematch.append(0)
        resrangematch.append(0)
        resrangematch.append(0)
    return resrangematch


#Performs calculations for each layer in a loop
#Returns dictionary in format key:"Layer N Target Thickness" value:[resrange, exit energy, entLET, extLET]
# i.e. Layer 1 Aluminum Thickness:[resrange, exit energy, entLET, extLET]
def runCalculations(energylist, energyin, ion, targetlist, thicklist, layercount):
    energyvals = getEnergyIndex(energylist, energyin)
    calcdict = {}
    keylist = []
    for i in range(0, layercount):
        rangelist = getRange(ion, targetlist[i])
        dedxlist = getDeDx(ion, targetlist[i])
        thickness = thicklist[i]
        key = "Layer " + str(i+1) + " " + str(targetlist[i]) + " " + str(thickness) + "mm"
        calcdict[key] = []
        keylist.append(key)
        if i == 0:
            index = energyvals[0]
            resrange = (rangelist[index] + (energyvals[2]-energyvals[1])*(rangelist[index+1]-rangelist[index])/(energylist[index+1]-energylist[index]))-thickness
            if resrange<0:
                resrange = 0
            calcdict[key].append(round(resrange,2))
            if resrange != 0:
                rangeindexlist = getRangeIndex(rangelist, resrange)
                rangeindex = rangeindexlist[0]
                closestenergy = energylist[rangeindex]
                energyup = energylist[rangeindex+1]
            if resrange>0:
                exitenergy = closestenergy + ((resrange-rangeindexlist[1])*(energyup - closestenergy) / (rangeindexlist[2] - rangeindexlist[1]))
            elif resrange==0:
                exitenergy = 0
            calcdict[key].append(round(exitenergy,2))
            entlet = dedxlist[index] + (energyvals[2] - energyvals[1]) * (dedxlist[index+1]-dedxlist[index]) / (energylist[index+1] - energyvals[1])
            calcdict[key].append(round(entlet,2))
            if resrange>0:
                extlet = dedxlist[rangeindex] + (exitenergy-energylist[rangeindex])*(dedxlist[rangeindex+1]-dedxlist[rangeindex])/(energylist[rangeindex+1]-energylist[rangeindex])
            elif resrange == 0:
                extlet = 0
            calcdict[key].append(round(extlet,2))
        elif i>0:
            if exitenergy!=0:
                energyvals = getEnergyIndex(energylist, exitenergy)
                exitenergy = energyvals[2]
                energyindex = energyvals[0]
                resrange = (rangelist[energyindex] + (exitenergy - energyvals[1])*(rangelist[energyindex+1]-rangelist[energyindex])/(energylist[energyindex+1]-energyvals[1])) - thickness
                if resrange<0:
                    resrange = 0
                rangeindexlist = getRangeIndex(rangelist, resrange)
                rangeindex = rangeindexlist[0]
                closestenergy = energy[rangeindex]
                energyup = energy[rangeindex+1]
                exitenergyprev = energyvals[2]
            elif exitenergy == 0:
                exitenergy = 0
                resrange = 0
                exitenergyprev = 0
            if resrange>0:
                exitenergy = closestenergy + ((resrange-rangeindexlist[1])*(energyup-closestenergy)/(rangeindexlist[2]-rangeindexlist[1]))
            elif resrange==0:
                resrange = 0
                exitenergy = 0
            calcdict[key].append(round(resrange,2))
            calcdict[key].append(round(exitenergy,2))
            if exitenergyprev !=0:
                entlet = dedxlist[energyindex] + (exitenergyprev-energylist[energyindex])*(dedxlist[energyindex+1]-dedxlist[energyindex])/(energylist[energyindex+1]-energylist[energyindex])
            elif exitenergyprev == 0:
                entlet = 0
            calcdict[key].append(round(entlet,2))
            if resrange>0:
                extlet = dedxlist[rangeindex] + (exitenergy-energylist[rangeindex])*(dedxlist[rangeindex+1]-dedxlist[rangeindex]) / (energylist[rangeindex+1]-energylist[rangeindex])
            elif resrange==0:
                extlet = 0
            calcdict[key].append(round(extlet,2))
    return calcdict

if cal_choice == 'Single Layer Calculator':
    dedx = getDeDx(ion, target)
    energy = getEnergy(ion)
    latstrag = getLatStrag(ion, target)
    longstrag = getLongStrag(ion, target)
    range_mm = getRange(ion, target)
    closestenergy = findInEnergy(energy)
    closestrange = findInRange(range_mm)
    closestdedx = findInDedx(dedx)
    diff = dedxdiff(closestdedx, dedx)

    # Energy Input Calcuations
    finddedx = (dedx[closestenergy[0]]) + (closestenergy[2] - closestenergy[1]) * (
            dedx[closestenergy[0] + 1] - dedx[closestenergy[0]]) / (energy[closestenergy[0] + 1] - closestenergy[1])
    findRange = (range_mm[closestenergy[0]]) + (closestenergy[2] - closestenergy[1]) * (
            range_mm[closestenergy[0] + 1] - range_mm[closestenergy[0]]) / (energy[closestenergy[0] + 1] - closestenergy[1])
    findxystrag = (latstrag[closestenergy[0]]) + (closestenergy[2] - closestenergy[1]) * (
            latstrag[closestenergy[0] + 1] - latstrag[closestenergy[0]]) / (
                          energy[closestenergy[0] + 1] - closestenergy[1])
    findzstrag = (longstrag[closestenergy[0]]) + (closestenergy[2] - closestenergy[1]) * (
            longstrag[closestenergy[0] + 1] - longstrag[closestenergy[0]]) / (
                         energy[closestenergy[0] + 1] - closestenergy[1])

    # Range Input Calculations
    findrenergy = (energy[closestrange[0]]) + (closestrange[2] - closestrange[1]) * (
            energy[closestrange[0] + 1] - energy[closestrange[0]]) / (range_mm[closestrange[0] + 1] - closestrange[1])
    findrdedx = (dedx[closestrange[0]]) + (closestrange[2] - closestrange[1]) * (
            dedx[closestrange[0] + 1] - dedx[closestrange[0]]) / (range_mm[closestrange[0] + 1] - closestrange[1])
    findrzstrag = (longstrag[closestrange[0]]) + (closestrange[2] - closestrange[1]) * (
            longstrag[closestrange[0] + 1] - longstrag[closestrange[0]]) / (
                          range_mm[closestrange[0] + 1] - closestrange[1])
    findrxystrag = (latstrag[closestrange[0]]) + (closestrange[2] - closestrange[1]) * (
            latstrag[closestrange[0] + 1] - latstrag[closestrange[0]]) / (range_mm[closestrange[0] + 1] - closestrange[1])

    # dE/dx Input Calculations
    finddxr = range_mm[diff[1]] + (closestdedx[2] - dedx[diff[1]]) * (range_mm[diff[1] + 1] - range_mm[diff[1]]) / (
            dedx[diff[1] + 1] - dedx[diff[1]])
    finddxenergy = energy[diff[1]] + (closestdedx[2] - dedx[diff[1]]) * (energy[diff[1] + 1] - energy[diff[1]]) / (
            dedx[diff[1] + 1] - dedx[diff[1]])
    finddxzstrag = longstrag[diff[1]] + (closestdedx[2] - dedx[diff[1]]) * (longstrag[diff[1] + 1] - longstrag[diff[1]]) / (
            dedx[diff[1] + 1] - dedx[diff[1]])
    finddxxystrag = latstrag[diff[1]] + (closestdedx[2] - dedx[diff[1]]) * (latstrag[diff[1] + 1] - latstrag[diff[1]]) / (
            dedx[diff[1] + 1] - dedx[diff[1]])

# Multi-Layer Calculator
else:
    energy = getEnergy(ion)
    closestenergy = findInEnergy(energy)
    layerdict = setLayers(input_layercount)
    targetlist = []
    thicklist = []
    for key in layerdict:
        targetlist.append(layerdict[key][0])
        thicklist.append(layerdict[key][1])
    calculations = runCalculations(energy, closestenergy[2], ion, targetlist, thicklist, 10)
    results = {}
    i=1
    for key in calculations:
        results["exit_energy_" + str(i)] = calculations[key][1]
        results["residual_range_" + str(i)] = calculations[key][0]
        results["entrance_LET_" + str(i)] = calculations[key][2]
        results["exit_LET_" + str(i)] = calculations[key][3]
        i=i+1
```
